```python
def retrieving_datasets(api):
    try:
        dataset_list = api.dataset_list()
        print("Please see a few datasets:")
        for dataset in dataset_list:  # Assuming dataset is an object with a 'title' attribute
            print(dataset.title)
    except Exception as e:
        print("Authentication failed:", e)  # Proper exception handling
    try:
        response = requests.get(api.dataset_list())  # Assuming this is correct
        if response.status_code == 403:
            print("Failed to fetch dataset file:", response.status_code)
    except Exception as e:
        print("Error during API request:", e)
```
